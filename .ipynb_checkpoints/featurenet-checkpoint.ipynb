```python
model_file = r'.\model\model.h5'
root_folder = r'.\data\pigeons'
target_folder = os.path.join(root_folder, '@target') 
full_image_file = os.path.join(root_folder, 'pigeons.png')
feature_file = os.path.join(root_folder, 'single_pigeon_small.png')
print("Feature {feature_file}")
Image(filename=feature_file)

ensure_dir(target_folder)
```

```python
feature_encoding = img_to_encoding(feature_file, FRmodel)
```

```python
print(f"feature file: {feature_file}")
print(f"full_image_file: {full_image_file}")
model_search_step = 20
best_image_file, sorted_distances = calculate_score_for_full_image_2(FRmodel, feature_file, full_image_file, model_search_step, target_folder, print_all_distances=False)
all_scores = sorted_distances
```

```python
img_feature = mpimg.imread(feature_file)
img_full = mpimg.imread(full_image_file)
w = 96
h = 96
crop_width = 96
crop_height = 96
def pictures(idx):
    x = all_scores[idx][1]
    y = all_scores[idx][0]
    score = all_scores[idx][2]
    
    pt1 = (x, y)
    pt2 = (x+w, y+h)
    radius = 50
    color = (127,255,0)
    idx_image = img_full.copy()
    cv2.rectangle(idx_image, pt1, pt2, color, thickness=5, lineType=8, shift=0)    
    print(f'The score of the {idx} ({x}, {y}) is {all_scores[idx][2]}')
    f, axarr = plt.subplots(1, 2, sharey=True)
    img = img_full[y:y+crop_height, x:x+crop_width]
    axarr[0].imshow(img_feature)
    axarr[1].imshow(img)
    plt.show()
    
    f, axarr = plt.subplots(1, 1, sharey=True)    
    axarr.imshow(idx_image)
    plt.show()
```

```python
sorted_scores = all_scores
sorted_scores.sort(key=lambda tup: tup[2])
print(f"number of scores: {len(sorted_scores)}")
interact(pictures, idx=(0,len(sorted_scores),1));
```
